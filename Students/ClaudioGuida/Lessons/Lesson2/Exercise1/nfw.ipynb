```python
import sys

sys.path.append('../../')

from pyACC.calculus import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Navarro-Frenk-White profile descibes the density of dark matter haloes
Rs=1 #scale radius
rho_0=1 #characteristic density
radius = np.linspace(1e-5, 100, 10000) #radius
rho = [rho_0/((r/Rs)*(1+(r/Rs))**2) for r in radius] #rho/rho_0

#enclosed mass
func = lambda r: 4*np.pi*(r/Rs)**2 * rho_0/((r/Rs)*(1+(r/Rs))**2)
M = [Integrate.adaptive_quadrature(func, 1e-10, r) for r in radius]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
ax1.loglog(radius/Rs, rho)
ax1.set_xlabel('r/Rs')
ax1.set_ylabel(r'$\rho/\rho_0$')
ax1.set_title('NFW profile')

ax2.loglog(radius/Rs, M)
ax2.set_xlabel('r/Rs')  
ax2.set_ylabel(r'$M/M_{sun}$')
ax2.set_title('Enclosed mass')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append("../")
from pyACC.Calculus import *
from pyACC.Cosmology import *
```

I fix M1 = 0.45Ms and M2=0.8Ms: orange stars

```python
Ms = 2e30 #kg
M1 = 0.45  #*Ms
M2 = 0.8   #*Ms
M = np.arange(M1,M2,0.001)
```

```python
N = []
for value in M:
    temp = lambda x : psi(x)
    N = np.append(N,integral(temp,M1,value))
#print('The number of stars between M1=0.45Ms and M2=0.8Ms is ', N)
plt.plot(M, N)
plt.xlabel('M')
plt.title('Total number of stars IMF')
```

```python
M_tot = []
for value in M:
    temp = lambda x : x*psi(x)
    M_tot = np.append(M_tot,integral(temp,M1,value))
#print('The number of stars between M1=0.45Ms and M2=0.8Ms is ', N)
plt.plot(M, M_tot)
plt.xlabel('M')
plt.title('Total mass')
#print('The total mass in stars born with mass between M1=0.45Ms and M2=0.8Ms is ', M_tot)
```

I change alpha

```python
N1 = []
N2 = []
for value in M:
    temp1 = lambda x : psi(x,alpha=2)
    N1 = np.append(N1,integral(temp1,M1,value))
    temp2 = lambda x : psi(x,alpha=2.55)
    N2 = np.append(N2,integral(temp2,M1,value))
#print('The number of stars between M1=0.45Ms and M2=0.8Ms is ', N)
plt.plot(M, N)
plt.plot(M, N1)
plt.plot(M, N2)
plt.legend(['alpha=2.35','alpha=2','alpha=2.55'])
plt.xlabel('M')
plt.title('Total number of stars IMF')
```

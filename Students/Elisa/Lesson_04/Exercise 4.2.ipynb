```python
z_f = 1
redshift = np.linspace(0,z_f,100)
Dc_t = [Dc(z) for z in redshift]    
```

```python
func = Interpolation(Dc,0,z_f)
Dc_i = func.lin_int()
```

```python
x = np.linspace(0,z_f,1000)
y = [Dc(z) for z in x]
temp1 = Dc_i(x)
```

```python
plt.plot(redshift,Dc_t)
plt.plot(x,temp1,'--g')
plt.xlabel('redshift')
plt.title('Comoving Distance')
plt.legend(['theoretical', 'linear interpolated'])
```

```python
x_err = np.linspace(0,1,1000)
temp_t = [Dc(z) for z in x_err]
deltad = func.deltad_lin()
plt.plot(x_err,deltad)
plt.xlabel('redshift')
plt.title('Residuals with linear intepolation')
```

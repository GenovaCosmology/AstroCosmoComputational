```python
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append("../")
from pyACC.Calculus import *
from pyACC.Cosmology import *
```

The Navarro-Frenk-White (NFW) profile is a model that describes the density profile of dark matter haloes

```python
def rho_DM(R,rho_0=1.4e7,rs=16.1):
    return rho_0/((R/rs)*(1+R/rs)**2)
```

```python
radius = np.linspace(1,100,1000) #kpc = galaxy dimension
n = radius.shape
M1 = []
```

```python
for value in radius:
    rho = lambda r : r*r*rho_DM(r)
    temp = integral(rho,0,value)
    M1 = np.append(M1,4*np.pi*temp)
```

```python
plt.plot(radius, M1)
plt.xlabel('radius [kpc]')
#plt.title('enclosed mass for $\rho_0 =1.4e7 M_s/kpc^3 and r_s=16.1 kpc$')
plt.title('Enclosed Mass')
```

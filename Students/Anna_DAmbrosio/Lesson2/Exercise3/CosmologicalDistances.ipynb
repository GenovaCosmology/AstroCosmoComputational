```python
# array of redshifts
z_list = np.linspace(0.0001, 2.5, 100)

# Calculate the comoving distance at each redshift in z_list
comoving_distance_res = (z_list)

# Calculate the angular diameter distance at each redshift in z_list
angular_diameter_distance_res = angular_diameter_distance(z_list)

# Calculate the luminosity distance at each redshift in z_list
luminosity_distance_res = luminosity_distance(z_list)

# Calculate the distance modulus at each redshift in z_list
distance_modulus_res = distance_modulus(z_list)
```

```python
plt.plot(z_list, comoving_distance_res, lw=3, color="k", label="Comoving distance")

plt.legend()
plt.xlabel("Redshift")
plt.ylabel("Distance [Mpc]");
```

```python
plt.plot(z_list, distance_modulus_res, lw=3, color="k", label="Distance modulus")
plt.legend()
plt.xlabel("Redshift")
plt.ylabel("$\mu$ [mag]");
```

```python
import sys
import numpy as np
import matplotlib.pyplot as plt

# Complete with appropriate path
sys.path.append("../")

# Complete with appropriate object to import
from pyACC.cosmology import flat_comoving_dist
from pyACC.calculus import quad_integral
```

```python
def Hz(z,Om,H0):
    Hz = H0*(np.sqrt(Om*(1+z)**3+(1-Om)))
    return Hz 
```

```python
H0=67. #km/Mpc/s
Om=0.3
print(H0,Om)
```

```python
redshift=np.linspace(0.001,2.5,1000)

Dc=[flat_comoving_dist(z,Om,H0) for z in redshift]
Da=(1/(1+redshift))*Dc
Dl=(1+redshift)*Dc

distance_modulus=5*np.log10(Dl*1e+6)-5
```

```python
plt.plot(redshift,Dc, label='comoving distance')
plt.plot(redshift,Da, label='angular diameter distance')
plt.plot(redshift,Dl, label='luminosity distance')
plt.legend()
plt.xlabel("Redshift")
plt.ylabel("Distance [Mpc]");
plt.show()
```

```python
plt.plot(redshift,distance_modulus, label='distance modulus')
plt.xlabel("Redshift")
plt.ylabel("Distance modulus")
```

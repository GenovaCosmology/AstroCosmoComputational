IMF: how mass stars is distributed

After implementing the IMF, I compute: the total number of stars, the average star mass and the average luminosity

I fix M1 = 0.8Ms and M2=1.4Ms: white stars

```python
Ms = 2e30 #kg
M1 = 0.8  #*Ms
M2 = 1.04   #*Ms
M = np.arange(M1,M2,0.001)
```

```python
#number of stars
def psi(m,psi_0=1,alpha=2.35):
    return psi_0*np.power(m,-alpha)

#psi_0 is constant, I fix it to 1
```

```python
#the total number of stars
N = []
for value in M:
    temp = lambda x : psi(x)
    N = np.append(N,integral(temp,M1,value))
#print('The number of stars between M1=0.45Ms and M2=0.8Ms is ', N)
plt.plot(M, N)
plt.xlabel('M [Ms]')
plt.title('Total number of stars IMF')
```

```python
#total mass
M_tot = []
for value in M:
    temp = lambda x : x*psi(x)
    M_tot = np.append(M_tot,integral(temp,M1,value))
#print('The number of stars between M1=0.45Ms and M2=0.8Ms is ', N)
plt.plot(M, M_tot)
plt.xlabel('M [Ms]')
plt.title('Total mass')
#print('The total mass in stars born with mass between M1=0.45Ms and M2=0.8Ms is ', M_tot)
```

```python
L = np.power(M_tot,4.7)
plt.plot(M, L)
plt.xlabel('M [Ms]')
plt.title('Luminosity')
```

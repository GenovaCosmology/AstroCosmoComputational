# 3D interpolation of distance modulus as a function of z,H0,omegam

Quantify residuals as in exercise #1 and test different interpolation techniques

The three parameters are indipendent so I can make three 1D interpolations instead of a 3D one

```python
z_f = 1
redshift = np.linspace(0.001,z_f,100)
int_H0 = np.arange(65,70,0.01)
int_omegam = np.arange(0.2,0.4,0.001)
```

# 3D interpolation of distance modulus as a function of z

```python
Dm_t = [Dm(z) for z in redshift]   
```

```python
func = Interpolation(Dm,0,z_f)
Dm_i = func.lin_int()
x = np.linspace(0,z_f,1000)
y = [Dm(z) for z in x]
temp1 = Dm_i(x)
```

```python
plt.plot(redshift,Dm_t)
plt.plot(x,temp1,'--g')
plt.xlabel('redshift')
plt.title('Distance modulus')
plt.legend(['theoretical', 'linear interpolated'])
```

```python
x_err = np.linspace(0,1,1000)
temp_t = [Dm(z) for z in x_err]
deltad = func.deltad_lin()
plt.plot(x_err,deltad)
plt.xlabel('redshift')
plt.title('Residuals with linear intepolation')
```

```python
Dm_i = func.spline()
temp2 = Dm_i(x)
```

```python
plt.plot(redshift,Dm_t)
plt.plot(x,temp2,'--g')
plt.xlabel('redshift')
plt.title('Comoving Distance')
plt.legend(['theoretical', 'linear interpolated'])
```

```python
x_err = np.linspace(0,1,1000)
temp_t = [Dm(z) for z in x_err]
deltad = func.deltad_spl()
plt.plot(x_err,deltad)
plt.xlabel('redshift')
plt.title('Residuals with linear intepolation')
```

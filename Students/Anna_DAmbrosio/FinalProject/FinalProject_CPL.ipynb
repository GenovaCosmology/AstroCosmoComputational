### DESI cosmology

```python
# definition of DESI cosmology

O_m_DESI = 0.295
w_de_DESI = -0.99
H_0_DESI = 68.52

w_0_DESI = -0.725
w_a_DESI = -1.06
```

### DESI data extraction

```python
# reading of the data and their management through the class dataDESImanager

data = open('DESIdata.txt', 'r')
lines = data.readlines()
lines = np.array(lines)
n = lines.size
DATA = np.zeros((n, n))
DATA = np.loadtxt("DESIdata.txt")

dataDESI, covDESI = dataDESImanager(DATA)

deltaDESI = np.sqrt(np.diag(covDESI))

z_DESI = dataDESI[:,0]
DV_DESI = dataDESI[:,1]
```

### PART 1: Test with fake data from a known cosmology

$\textbf{C1 cosmology}$: \
$H_0 = 68.52$ \
$\Omega_m = 0.295$ \
$\Omega_{\gamma} = 0.0$ \
$\Omega_{\nu} = 0.0$ \
$\Omega_{\Lambda} = 0.705$ \
$w_{\Lambda} = -0.99$ \
$w_0 = -0.725$ \
$w_a = -1.06$

```python
# definition of C1 cosmology

O_m_C1 = O_m_DESI
w_de_C1 = w_de_DESI
H_0_C1 = H_0_DESI
w_0_C1 = w_0_DESI
w_a_C1 = w_a_DESI
```

```python
# generation of fake data given a gaussian with mean the value calculated with the given codmology and error taken from DESI data

mean_DV_C1 = distance_volume_over_rd(z=z_DESI, H_0=H_0_C1, O_m=O_m_C1, O_gam=0, O_nu=0, O_de=(1-O_m_C1), w_de=w_de_C1, w0=w_0_C1, wa=w_a_C1)

data_C1 = np.random.normal(mean_DV_C1, deltaDESI)
```

```python
# creation of data set with DESI format

zero_column = np.zeros(z_DESI.size)

DATA_C1 = np.column_stack([z_DESI, zero_column, zero_column, zero_column, zero_column, data_C1, deltaDESI]) 
```

```python
# plot of the data

fig, ax = plt.subplots()
ax.errorbar(DATA_C1[:,0], DATA_C1[:,5], yerr=DATA_C1[:,6], fmt='o', ms=5, label='C1 data', color='green')
z_array = np.linspace(z_DESI[0], z_DESI[-1], 1000)
ax.set_xlabel(r'$z$')
ax.set_ylabel(r'$D_V$')

D_V_th = distance_volume_over_rd(z_array, H_0=68.52, O_m=O_m_C1, O_gam=0, O_nu=0, O_de=(1-O_m_C1), w_de=w_de_C1, w0=w_0_C1, wa=w_a_C1)
ax.plot(z_array, D_V_th, label=r'$D_V$ model', ls='--', color='black')

ax.grid()
ax.legend()
```

```python
# definition of function that describes the data

func_dist = lambda z, w0,wa : distance_volume_over_rd(z, H_0=H_0_C1, O_m=O_m_C1, O_gam=0, O_nu=0, O_de=(1-O_m_C1), w_de=w_de_C1, w0=w0, wa=wa)
```

```python
# calculation of logLikelihood with flat priors

infs=np.array([-1.5,-2])
sups=np.array([-0.2,-0.5])
LogL_C1 = logL_flatPrior(DATA_C1, func_dist=func_dist, infs=infs, sups=sups)
```

```python
# sampling with MCMC

w0_point = w_0_C1
wa_point = w_a_C1
point0 = np.array([w0_point,wa_point])
ds = np.array([0.01,0.01])
covmat = np.diag(ds**2)
N = 100000

results_C1, eff_results_C1 = sampler.Metropolis_Hastings(LogL_C1, x0=point0, N=N, cov=covmat, type='gaussian', log=True)
print(covmat)
```

```python
# corner plot

w0_samples_C1 = results_C1[:,0]
wa_samples_C1 = results_C1[:,1]

names_C1 = ['w_0', 'w_a']
labels_C1 = ['w_0', 'w_a']

samples_C1 = getdist.mcsamples.MCSamples(samples=results_C1, names=names_C1, labels=labels_C1, label='C1')

g = plots.get_subplot_plotter()
g = plots.get_subplot_plotter(subplot_size=3)
g.settings.figure_legend_frame = False
g.settings.alpha_filled_add=0.4
g.settings.title_limit_fontsize = 14

g.triangle_plot(samples_C1, legend_labels=['C1'], 
                filled=True,
                line_args={'ls':'--','lw':1.5, 'color':'green'},
                contour_colors=['green'], 
                legend_loc='upper right',
                title_limit=1, # first title limit (for 1D plots) is 68% by default
                markers={'w_0':w_0_C1, 'w_a':w_a_C1},)
```

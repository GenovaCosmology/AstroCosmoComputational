```python
# define a function that would give you a 2D distribution. Do not pick a Gaussian.
def fun(x,y):
    return (x**2+y**2)*np.exp(-(np.sqrt(x)+np.sqrt(y)))

# define a prior box:
x_min = 0.0
x_max = 10.0
y_min = 0.0
y_max = 10.0
steps = 1000

x_ar = np.linspace(x_min, x_max,steps)
y_ar = np.linspace(y_min, y_max,steps)
```

```python
# sample your reference probability on a 2D regular grid in the prior box
x_grid, y_grid = np.meshgrid(x_ar, y_ar)
```

```python
# normalize the pdf
norm = integrate.nquad(fun,[[x_min,x_max],[y_min,y_max]])[0]

def fun_norm(x, y, norma):
    return (1/norma)*fun(x,y)

pdf = fun_norm(x_grid, y_grid, norm)
```

```python
# contour plot the distribution (on generic iso-contours, the default ones work)
fig, ax = plt.subplots()
CS = ax.contourf(x_grid, y_grid, pdf, levels=10)
CS2 = ax.contour(CS, levels=CS.levels[::2], colors='black')

cbar = fig.colorbar(CS)
cbar.add_lines(CS2)
```

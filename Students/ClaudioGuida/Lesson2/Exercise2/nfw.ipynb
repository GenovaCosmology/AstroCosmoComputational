```python
import sys

sys.path.append('../../')

from pyACC.calculus import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Navarro-Frenk-White profile descibes the density of dark matter haloes
def NFW(r, rho0, rs):
    return rho0 / (r/rs * (1 + r/rs)**2)
```

```python
Rs=1 #scale radius
rho_0=1 #characteristic density
r_min = 1e-6
r_max = 1e4
radius = np.linspace(r_min, r_max, 10000) #radius in Kpc
rho = [NFW(r,rho_0,Rs) for r in radius] #rho/rho_0

#enclosed mass
func = lambda r: 4*np.pi*(r/Rs)**2/((r/Rs)*(1+(r/Rs))**2)
M = [Integrate.adaptive_quadrature(func, 1e-10, r) for r in radius]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
ax1.loglog(radius, rho)
ax1.set_xlabel('r (Kpc)')
ax1.set_ylabel(r'$\rho/\rho_0$')
ax1.set_title('NFW profile')

ax2.loglog(radius, M)
ax2.set_xlabel('r (Kpc)')  
ax2.set_ylabel(r'$M/M_{sun}$')
ax2.set_title('Enclosed mass')
plt.show()
```

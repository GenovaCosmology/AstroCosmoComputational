```python
# array of redshifts
z_list = np.arange(0.0, 2.0, 0.01)

# Calculate the comoving distance at each redshift in z_list with LCDM model
comoving_distance_LCDM = comoving_distance(z_list,H_0=67.0, O_m=0.319, O_gam=0, O_nu=0, O_de = 0.681, w_de=-1)

# Calculate the comoving distance at each redshift in z_list with wCDM model
comoving_distance_wCDM = comoving_distance(z_list,H_0=67.0, O_m=0.319, O_gam=0, O_nu=0, O_de = 0.681, w_de=-1.1)
```

```python
zeros = np.zeros((z_list.size,3))
zeros[:,0] = z_list
zeros[:,1] = comoving_distance_LCDM
zeros[:,2] = comoving_distance_wCDM
title = np.array(["z", "DC_LCDM", "DC_wCDM"])
data = np.vstack((title,zeros))
```

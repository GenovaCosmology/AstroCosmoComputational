```python
import sys
import matplotlib.pyplot as plt
import scipy
from scipy.stats import norm

# Complete with appropriate path
sys.path.append("../")

# Complete with appropriate object to import
from pyACC.sampling import rejection
```

```python
# Define a target distribution yuou want to sample from
# generate random numbers from N(0,1)
target_normal = norm.rvs(size=10000,loc=5,scale=2)

# plot the histogram of the target distribution
plt.hist(target_normal, bins=100, density=True, alpha=0.6, color='g');
```

```python
# Choose a proposal distribution that is easy to sample from.
proposal_normal = norm.rvs(size=10000,loc=0,scale=1)

# plot the histogram of the proposal distribution
plt.hist(proposal_normal, bins=100, density=True, alpha=0.6, color='r');
```

```python
# sample from the target distribution using the rejection sampling algorithm
samples = rejection(target_normal, proposal_normal,0.001, 0, 1, 100000)
```

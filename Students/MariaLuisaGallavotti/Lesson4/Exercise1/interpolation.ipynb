```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 8, 27, 64]

interp = interp1d(x, np.array(x)**3, kind='linear')

d=0.1
x_new = np.arange(0,4+d,d)
y_new = interp(x_new)
print(y,y_new)
print(len(y),len(y_new))
```

```python
diff = []
'''
for i, x_val in enumerate(x_new):
    y_new_index=int(i*1/d)
    diff.append(x_val**3 - y_new[y_new_index])
'''
for i in range(0,len(x_new)):
    diff.append(x_new[i]**3 - y_new[i])
    print(diff)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x_new, y_new, 'bo', label='Interpolated point')
plt.plot(x, y, 'ro', label='Original data')
plt.plot(x, interp(x), 'g-', label='Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Interpolation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(x_new, np.abs(diff), label='Difference (y - y_new)')
plt.xlabel('x')
plt.ylabel('Difference')
plt.title('Difference between y and interpolated y')
plt.axhline(0, color='r', linestyle='--', linewidth=1)
plt.legend()
plt.grid(True)
plt.show()
```

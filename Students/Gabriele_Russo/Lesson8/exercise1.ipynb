```python
#Usefull Imports
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Grid limits
Delta_X = np.linspace(0, 1, 10)
Delta_Y = np.linspace(0, 1, 10)
Delta_Z = np.linspace(0, 1, 10)

delta_x = Delta_X[1] - Delta_X[0]
print(delta_x)
```

```python
#Grid creation
X, Y, Z = np.meshgrid(Delta_X, Delta_Y, Delta_Z)
```

```python
#Giving all points the same uniform density
delta = 5 #imposing constant value

uniform_data = np.full(X.shape, delta)
```

```python
# New figure
fig = plt.figure()

# 3D subplot to figure
ax = fig.add_subplot(111, projection='3d')

# Plotting uniform data
sc = ax.scatter(X, Y, Z, c=uniform_data, cmap='viridis')

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.colorbar(sc, label='Color')

plt.show()
```

```python
#Fast Fourier Transformation of the grid 
uniform_data_transform = np.fft.rfftn(uniform_data)
print(uniform_data_transform.shape)
#print(uniform_data_transform)
```

```python
# Shape of the 3D grid
shape = uniform_data_transform.shape

# Sampling frequencies along each axis in the Fourier space
freq_x = np.fft.fftfreq(shape[0], delta_x)
freq_y = np.fft.fftfreq(shape[1], delta_x)
freq_z = np.fft.fftfreq(shape[2], delta_x)

# Create coordinate grids in Fourier space
X_fourier, Y_fourier, Z_fourier = np.meshgrid(freq_x, freq_y, freq_z, indexing='ij')

# Now X_fourier, Y_fourier, and Z_fourier contain the corresponding coordinates in Fourier space
```

```python
# New figure
fig = plt.figure()

# 3D subplot to figure
ax = fig.add_subplot(111, projection='3d')

# Plotting uniform data
sc = ax.scatter(X_fourier, Y_fourier, Z_fourier, c=uniform_data_transform, cmap='viridis')

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.colorbar(sc, label='Color')

plt.show()
```

```python
# Compute the squared magnitude of the Fourier coefficients
power_spectrum = np.abs(uniform_data_transform)**2

#Normalize the power spectrum ensuring its values represent the power per unit area in the frequency domain
power_spectrum /= np.prod(uniform_data.shape)

# Print the shape of the power spectrum
print(power_spectrum.shape)

print(power_spectrum)
```

```python
# New figure for power spectrum
fig = plt.figure()

# 3D subplot to figure
ax = fig.add_subplot(111, projection='3d')

# Plotting power spectrum
X, Y, Z = np.meshgrid(range(power_spectrum.shape[0]), range(power_spectrum.shape[1]), range(power_spectrum.shape[2]), indexing='ij')
'''
Note that:
- range(power_spectrum.shape[0]), range(power_spectrum.shape[1]) and range(power_spectrum.shape[2]) generate 1D coordinate vectors 
along the X, Y, and Z dimensions respectively.
These ranges correspond to the shape of the power spectrum array;
- indexing='ij' specifies that the output coordinate matrices should use the 'ij' indexing convention, 
meaning the first axis corresponds to rows and the second axis corresponds to columns;
- the function returns three 3D coordinate matrices: X, Y, and Z, where each matrix contains the corresponding coordinates 
for each point in the 3D space defined by the power spectrum array.
'''
sc = ax.scatter(X, Y, Z, c=power_spectrum.flatten(), cmap='viridis')
'''
Note that: 
- the .flatten() method converts the 3D power spectrum array into a 1D array to match the number of points.
'''

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add color bar
plt.colorbar(sc, label='Power Spectrum')

plt.show()
```

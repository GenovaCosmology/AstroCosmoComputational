```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the differential equation dy/dt = f(y, t)
def model(y, t):
    k = 0.3  # Example parameter
    dydt = -k * y  # Example differential equation, e.g., for exponential decay
    return dydt

# Define initial condition
y0 = 5.0  # Example initial condition

# Define time points to solve for
t = np.linspace(0, 20, 100)  # Example time points

# Solve the differential equation
y = odeint(model, y0, t)

# Plot the results
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('Solution of the Differential Equation')
plt.grid(True)
plt.show()
```

```python
import sys

# Complete with appropriate path
sys.path.append("../../")

# Complete with appropriate object to import
from pyACC.calculus import *
from pyACC.cosmology import distances
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#setting costants
c = 299792 #c in  km/s
H_0 = 100 #h km /s/Mpc
D_h = c/H_0 #hubble distance in Mpc/h
```

```python
z=np.arange(0, 2, 0.01)+0.01/2
com_dist_lcdm=[distances.comoving_distance(redshift,H_0=67,omega_m=0.319,omega_r=0) for redshift in z]
com_dist_wcdm=[distances.comoving_distance(redshift,H_0=67,omega_m=0.319,omega_r=0,w=-1.1) for redshift in z]
```

```python
#for i in range(len(z)):
data = np.column_stack((z, com_dist_lcdm, com_dist_wcdm) )
#np.savetxt('../Lesson2/Exercise3/comoving_distance.txt', data , delimiter='\t', header='z\t  Dc_lcdm\t Dc_wcdm')
```

```python
plt.plot(z,com_dist_lcdm, label='lcdm')
plt.plot(z,com_dist_wcdm, label='wcdm')
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel('z')
plt.ylabel('Dc [Mpc/h]')
plt.legend()
```

```python
plt.plot(z, [cm/D_h for cm in com_dist_lcdm])
plt.xlabel('z')
plt.ylabel('Dc/Dh')
```

```python
#angular diameter distance
ang_dist=[distances.angular_distance(redshift) for redshift in z]
plt.plot(z,ang_dist)
plt.xlabel('z')
plt.ylabel('Da [Mpc/h]')    
```

```python
plt.plot(z, [da/D_h for da in ang_dist])
plt.xlabel('z')
plt.ylabel('Da/Dh')
```

```python
#luminosity distance
lum_dist = [distances.luminosity_distance(redshift) for redshift in z]
plt.plot(z, lum_dist)
plt.xlabel('z')
plt.ylabel('Dl  [Mpc/h]')  
```

```python
plt.plot(z, [dl/D_h for dl in lum_dist])
plt.xlabel('z')
plt.ylabel('Dl/Dh')
```

```python
dist_modulus = [distances.distance_modulus(redshift) for redshift in z]
plt.plot(z, dist_modulus)
plt.xlabel('z')
plt.ylabel('Dm')
```

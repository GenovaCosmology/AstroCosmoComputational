```python
# define a function that would give you a 2D distribution. Do not pick a Gaussian.
def f(x1,x2):
    return (x1)*(x2**3)*np.exp(-(x1+x2**2))

# define a prior box:
x1_min=0
x1_max=6
x2_min=0
x2_max=2.8
Nsteps=1000
X1=np.linspace(x1_min,x1_max,Nsteps)
X2=np.linspace(x2_min,x2_max,Nsteps)
X_1,X_2=np.meshgrid(X1,X2)

# sample your reference probability on a 2D regular grid in the prior box
f_prob=f(X_1,X_2)
plt.imshow(f_prob,extent=[x1_min,x1_max,x2_min,x2_max], cmap=cm.jet, origin='lower')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Not normalized probability distribution')
plt.colorbar()
```

```python
# normalize the pdf
norm=integrate.nquad(f, [[x1_min, x1_max], [x2_min, x2_max]])[0]
print(norm)
def PDF(x1,x2):
    return f(x1,x2)/norm
```

```python
# contour plot the distribution (on generic iso-contours, the default ones work)
PDF_grid=PDF(X_1,X_2)
fig,ax1=plt.subplots()
CS = ax1.contourf(X_1,X_2,PDF_grid, 20)
CS2 = ax1.contour(CS, levels=CS.levels[::2], colors='r',linewidths=0.5)
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_title('PDF contour plot')
cbar = fig.colorbar(CS)
cbar.ax.set_ylabel('PDF intensity')
cbar.add_lines(CS2)
```

```python
z_f = 1000
#redshift = np.linspace(0,z_f,1000)
redshift = np.logspace(-2,2,1000)
```

```python
plt.loglog(redshift,[Dc(z) for z in redshift])
plt.xlabel('log(z)')
plt.ylabel('log(Dc)')
plt.title('Comoving distance')
```

```python
plt.loglog(redshift,[Da(z) for z in redshift])
plt.xlabel('log(z)')
plt.ylabel('log(Da)')
plt.title('Angular diameter distance')
```

```python
plt.loglog(redshift,[Dl(z) for z in redshift])
plt.xlabel('log(z)')
plt.ylabel('log(Dl)')
plt.title('Luminosity distance')
```
